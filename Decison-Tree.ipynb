```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('salaries.csv')
df
```

```python
inputs = df.drop('salary_more_then_100k',axis = 'columns')
target = df['salary_more_then_100k'] 
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])

inputs
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis= 'columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,2,1]])
```

```python
model.predict([[2,0,1]])
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
inputs = df[['Pclass','Sex','Age','Fare',]]
mn = inputs['Age'].mean()
```

```python
target = df['Survived']
target
```

```python
inputs['Age'].fillna(mn,inplace =True)
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Sex = LabelEncoder()
```

```python
inputs['le_Sex'] = le_Sex.fit_transform(inputs['Sex'])
inputs
```

```python
finputs = inputs.drop(['Sex'],axis = 'columns')
finputs
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(finputs,target)
```

```python
model.score(finputs,target)
```

```python
model.predict([[3,35,8,1]])
```
